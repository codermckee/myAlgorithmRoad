```python
import numpy as np
from numpy import random
```

```python
a = random.random_integers(1,6)
b = random.random_integers(1,6)
x = random.randint(0,3,size=(a,b))
print x
matrix = x
location = []
for i in range(a):
    for j in range(b):
        if matrix[i][j] == 0:
            location.append((i,j))
print location
```

```python
def get_kill_num(mode,matrix,location):   
    kill_enermy = {}
    for i in location:
        #n = n+1
        count = 0
        x = i[0]
        y = i[1]
        if mode == 'down':
            while x+1<a and y<b and matrix[x+1][y]!=2:
                if matrix[x+1][y] == 1:
                    count += 1
                x = x+1
            kill_enermy[str(i[0])+','+str(i[1])]= count
        if mode == 'right':
            while x<a and y+1<b and matrix[x][y+1]!=2:
                if matrix[x][y+1] == 1:
                    count += 1
                y = y+1
            kill_enermy[str(i[0])+','+str(i[1])]= count
        if mode == 'left':
            while x>=0 and y-1>=0 and matrix[x][y-1]!=2:
                if matrix[x][y-1] == 1:
                    count += 1
                y = y-1
            kill_enermy[str(i[0])+','+str(i[1])]= count
        if mode == 'up':
            while x-1>=0 and y>=0  and matrix[x-1][y]!=2:
                if matrix[x-1][y] == 1:
                    count += 1
                x = x-1
            kill_enermy[str(i[0])+','+str(i[1])]= count
        
    return kill_enermy
```

```python
kill_enermy1 = {}
kill_enermy2 = {}
kill_enermy3 = {}
kill_enermy4 = {}
kill_enermy1 = get_kill_num('left',matrix,location)
kill_enermy2 = get_kill_num('up',matrix,location)
kill_enermy3 = get_kill_num('down',matrix,location)
kill_enermy4 = get_kill_num('right',matrix,location)
```

```python
kill_enermy = {}
for i in location:
    kill_enermy[str(i[0])+','+str(i[1])]=kill_enermy1[str(i[0])+','+str(i[1])]+kill_enermy2[str(i[0])+','+str(i[1])]+kill_enermy3[str(i[0])+','+str(i[1])]+kill_enermy4[str(i[0])+','+str(i[1])]
kill =  sorted(kill_enermy.iteritems(), key=lambda d:d[1], reverse = True)
print '坐标及杀死敌人数量为(TOP3)'
print kill[0],kill[1],kill[2]
```
